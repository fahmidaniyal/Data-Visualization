##Types of Data Visualization

```python
import pandas as pd

tips = pd.read_csv('tips.csv')
tips.head()
```

```python
tips.shape
```

```python
tips['tip'].max()
```

##Line Plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,5))
plt.plot(tips['tip'], marker = 'o', linestyle = '-.')
```

##Bar Chart

```python
x = tips['day']
y = tips['tip']

plt.bar(x,y)

plt.xlabel('Day')
plt.ylabel('Tips')
```

Kustomisasi yang tersedia untuk Bar Chart :
- color: Untuk wajah bar
- edgecolor: Warna tepi bilah
- linewidth: Lebar tepi bilah
- width: Lebar bilah

```python
plt.bar(x,y, color = 'yellow', edgecolor = 'black', linewidth = 2)
plt.xlabel('Day')
plt.ylabel('Tips')
```

##Histogram

syntax:

matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype=’bar’, align=’mid’, orientation=’vertical’, rwidth=None, log=False, color=None, label=None, stacked=False, \*, data=None, \*\*kwargs)

```python
plt.hist(tips['total_bill'])

plt.xlabel('Total Bill')
plt.ylabel('Frequency')
```

Kustomisasi yang tersedia untuk Histogram

- bins: Jumlah tempat sampah dengan lebar yang sama
- color: Untuk mengubah warna wajah
- edgecolor: Warna tepi
- linestyle: Untuk garis tepi
-  alpha: nilai campuran, antara 0 (transparan) dan 1 (buram)

```python
plt.hist(tips['total_bill'], bins = 25,
                             color = 'orange',
                             edgecolor = 'blue',
                             linestyle = '-.',
                             alpha = 0.75)

plt.xlabel('Total Bill')
plt.ylabel('Frequency')
```

##Scatter Plot

```python
x = tips['total_bill']
y = tips['tip']

plt.scatter(x,y)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

Grafik ini bisa lebih bermakna jika kita bisa menambahkan warna dan juga mengubah ukuran titik-titiknya. Kita dapat melakukan ini dengan menggunakan parameter c dan s masing-masing dari fungsi scatter.

Kustomisasi yang tersedia untuk scatter adalah –

- s: ukuran penanda (bisa skalar atau array dengan ukuran sama dengan ukuran x atau y)
- c: warna urutan warna untuk spidol
- marker: gaya penanda
- linewidths: lebar batas penanda
- edgecolor: warna batas penanda
- alpha: nilai campuran, antara 0 (transparan) dan 1 (buram)


Kita juga dapat menampilkan bilah warna menggunakan metode colorbar()

```python
plt.scatter(x, y,
            c = tips['size'],         #size value determined by color as shown in colorbar
            s = tips['total_bill'],   #total_bill value affect the marker's size
            marker = 'D',             #diamond marker shape
            alpha = 0.5)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.colorbar()
```

##Pie Chart

syntax : matplotlib.pyplot.pie(data, explode=None, labels=None, colors=None, autopct=None, shadow=False)

```python
cars = ['AVANZA','BENTLEY','CORVETTE','DACIA','EXAGON']
sale = [105,23,45,77,12]

plt.pie(sale, labels = cars)

plt.show()
```

Kustomisasi yang tersedia untuk pie chart adalah –

- explode: Memindahkan irisan plot
- autopct: Beri label baji dengan nilai numeriknya.
- color: Atribut digunakan untuk memberikan warna pada irisan.
- shadow: Digunakan untuk membuat bayangan baji.

```python
explode = [0.05, 0.2, 0, 0, 0.2]
colors = ('silver','white','blue','red','green')

plt.pie(sale,
        labels = cars,
        explode = explode,
        autopct = '%.2f%%',
        colors = colors,
        shadow = True)

plt.show()
```
